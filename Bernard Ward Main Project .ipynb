```python
#Importing Data data from a CSV file

#Print the data on screden to display the lis

import csv

car_data = pd.read_csv("cars2.csv")

#Print the data ( returns 5076 rows  18 columns)
car_data
```

```python
#Remove Columns not required (Data sorting. We are only interested in relevant data)
# We can see that there is lots of data that is not relevant So we will remove the irrelevant Columns
#use the .copy() method on the DataFrame to get a copy of it, then you can now drop the irrelevant columns from the new DataFrame.

car_data2 = car_data.copy()   # make a copy of the dataframe object
car_data2.drop(columns = ['Height', 'Hybrid','Year', 'ID' , 'Driveline', 'Engine Type','Length','Width', 'Transmission','Number of Forward Gears', 'Make'], inplace=True )   # drop the irrelevant columns
car_data2
```

```python
# example of sorting dataframe
# Here sorted by 'Horsepower' 
Car_data_final = car_data2.sort_values(by='Horsepower')
Car_data_final
```

```python
# Visualisation of Data
#using Matplot.lib

# Used !pip install matplotlib to install  Matplotlib and run comand returns "generated new fontManager"

import matplotlib.pyplot as plt
import seaborn as sns          
           
```

```python
# Proior to visualization of data we look at the aailable data types in the Data frame

#The following is returned
#City mpg           int64
#Fuel Type         object
#Highway mpg        int64
#Classification    object
#Model Year        object
#Horsepower         int64
#Torque             int64
#dtype: object

Car_data_final.dtypes
```

```python
#Visualise the chart with a Scatter plot of various comnbinations
#  Horsepower versus Highway mpg
# Horsepower versus City mpg


plt.scatter( Car_data_final ['Horsepower'],  Car_data_final  ['Highway mpg'], s=2 )

plt.title ('HorsePower versus Highway mpg')
plt.xlabel('Horsepower') 
plt.ylabel('Highway mpg')
```
